## YOLOv8 Training

```python
ultralytics.checks()
```

```python
from ultralytics import YOLO
# Load a pre-trained model 
model = YOLO('/kaggle/input/yolov8_beachlitter/keras/default/5/YOLOv8x_50epoch_batch16_LR0005_last.pt')

# Train the model 
results = model.train(data='/kaggle/working/data.yaml', epochs=50, batch = 16, imgsz=(736,544), device = [0,1] , workers = 8, pretrained = False, optimizer = 'SGD', lr0 = 0.005, momentum = 0.9) # argumen tinggal tambah
```

## Load model

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 segmentation model
model = YOLO("/kaggle/working/runs/segment/train/weights/best.pt")
```

## Image segmentation prediction

```python
from PIL import Image 
  
# get image 
img_path = "/kaggle/input/beachlitter-dataset-for-yolo/prepped_YOLO/test/images/002800.jpg"
img_temp = Image.open(img_path) 
  
# get width and height 
width = img_temp.width 
height = img_temp.height 

results = model.predict(img_path, save=True, imgsz=(height, width), conf=0.3, show_labels = False, show_conf = False, show_boxes = False)
```
